#### 11.	Add a column “Range” to the existing dataframe for below condition
a.	1 for Effective_Price > 500
b.	0.5 otherwise

```python
def calc_range_col(row):
    if row['Effective_Price'] > 500:
        return 1
    else:
        return 0.5
df['Range'] = df.apply(calc_range_col, axis=1)
df
```
